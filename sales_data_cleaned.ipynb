### 🧹 Data Cleaning and Preprocessing
Key cleaning steps performed:

- Converted date columns into proper datetime format for time-based analysis
- Removed duplicate records to maintain data integrity
- Handled missing values to avoid issues during analysis

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])
df = df.drop_duplicates()
df = df.fillna(0)
print("Data cleaned successfully!")
```

### 🧾 Column Name Standardization
Column names were cleaned and standardized to ensure consistency:
- Removed extra spaces
- Converted to lowercase
- Replaced special characters with underscores

This improves readability and avoids errors in analysis.

```python
df.columns = (
    df.columns
    .str.strip()
    .str.lower()
    .str.replace("-", "_")
)
```
